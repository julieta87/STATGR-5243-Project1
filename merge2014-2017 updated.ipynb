### Data Cleaning

```python
def preprocess_data(df):
    columns_to_keep = ["tpep_pickup_datetime", "tpep_dropoff_datetime", "passenger_count",
                       "trip_distance", "payment_type", "fare_amount", "tip_amount", "total_amount", "PULocationID", 
                       "DOLocationID"]
    
    df = df[columns_to_keep].copy()
    
    df["tpep_pickup_datetime"] = pd.to_datetime(df["tpep_pickup_datetime"])
    df["tpep_dropoff_datetime"] = pd.to_datetime(df["tpep_dropoff_datetime"])
    
    df = df[(df["fare_amount"] > 0) & (df["trip_distance"] > 0)]

    df["trip_duration"] = (df["tpep_dropoff_datetime"] - df["tpep_pickup_datetime"]).dt.total_seconds() / 60
    
    df = df[(df["trip_duration"] > 1) & (df["trip_duration"] < 300)]
    
    return df
```

### Feature Engineering

```python
daily_summary_features = daily_summary.copy()

daily_summary_features['pickup_date'] = pd.to_datetime(daily_summary_features['pickup_date'])

daily_summary_features['month'] = daily_summary_features['pickup_date'].dt.month
daily_summary_features['quarter'] = daily_summary_features['pickup_date'].dt.quarter

daily_summary_features['is_weekend'] = daily_summary_features['day of week'].isin(['Saturday', 'Sunday']).astype(int)

day_map = {'Monday':0, 'Tuesday':1, 'Wednesday':2, 'Thursday':3, 'Friday':4, 'Saturday':5, 'Sunday':6}
daily_summary_features['day_of_week_num'] = daily_summary_features['day of week'].map(day_map)

daily_summary_features['day_of_week_sin'] = np.sin(2 * np.pi * daily_summary_features['day_of_week_num'] / 7)
daily_summary_features['day_of_week_cos'] = np.cos(2 * np.pi * daily_summary_features['day_of_week_num'] / 7)
```

```python
daily_summary_features['fare_per_passenger'] = (
    daily_summary_features['total fares'] / daily_summary_features['sum of passengers']
)

five_boroughs = ["Manhattan", "Brooklyn", "Queens", "Bronx", "Staten Island"]

for borough in five_boroughs:
    col_name = borough.lower().replace(" ", "_")
    
    daily_summary_features[f"{col_name}_pickup_share"] = (
        daily_summary_features[f"{borough}_Pickups"] / daily_summary_features["frequency_of_rides"]
    )
    
    daily_summary_features[f"{col_name}_dropoff_share"] = (
        daily_summary_features[f"{borough}_Dropoffs"] / daily_summary_features["frequency_of_rides"]
    )
```

```python
daily_summary_features = daily_summary_features.sort_values('pickup_date')

daily_summary_features['total_fares_lag1'] = daily_summary_features['total fares'].shift(1)

daily_summary_features['total_fares_rolling7'] = daily_summary_features['total fares'].rolling(7).mean()

daily_summary_features['frequency_of_rides_rolling7_std'] = daily_summary_features['frequency_of_rides'].rolling(7).std()
```

```python
daily_summary_features['log_total_fares'] = np.log1p(daily_summary_features['total fares'])  # log(1 + x)

bins = [0, 10, 20, 100]
labels = ['Low', 'Medium', 'High']
daily_summary_features['tip_percent_bin'] = pd.cut(
    daily_summary_features['tip percentage'],
    bins=bins,
    labels=labels,
    include_lowest=True
)
```

```python
borough_pickup_cols = [
    'Bronx_Pickups','Brooklyn_Pickups','EWR_Pickups',
    'Manhattan_Pickups','Queens_Pickups','Staten Island_Pickups','Unknown_Pickups'
]
daily_summary_features['dominant_pickup_borough'] = daily_summary_features[borough_pickup_cols].idxmax(axis=1)

daily_summary_features['nonzero_pickup_boroughs'] = (
    daily_summary_features[borough_pickup_cols] > 0
).sum(axis=1)
```
