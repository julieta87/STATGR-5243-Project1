```python
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')
path = '/content/drive/MyDrive/Colab Notebooks/STAT 5243/'
```

```python
import dask.dataframe as dd
import glob

# Get all Parquet files
files = sorted(glob.glob(path+"yellow_tripdata_2016-*.parquet"))

sampled_dfs = []  # Store sampled data

for file in files:
    print(f"Processing: {file}")

    # Load one month's data
    df = dd.read_parquet(file, engine="pyarrow")

    # Take a random 10% sample
    df_sampled = df.sample(frac=0.1, random_state=42)  # Set random_state for reproducibility

    # Save the sampled dataset as CSV (can split into multiple files if large)
    sampled_file = file.replace(".parquet", "_sampled.csv")
    df_sampled.to_csv(sampled_file, index=False, single_file=True)
    print(f"Saved sampled dataset: {sampled_file}")

    sampled_dfs.append(df_sampled)

# Merge all sampled months into one dataset
final_df = dd.concat(sampled_dfs, axis=0)

# Save final merged dataset as CSV
final_df.to_csv("sampled_yellow_tripdata_2016.csv", index=False, single_file=True)
```
